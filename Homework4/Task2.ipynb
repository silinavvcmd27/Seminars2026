```python
from Bio import motifs
from Bio import SeqIO
from Bio.Seq import Seq

# 1. Создание мотива из последовательностей Задания 1
instances = [Seq(s) for s in sites]
m = motifs.create(instances)

# Рассчитываем PWM (матрицу весов)
pwm = m.counts.normalize(pseudocounts=0.5).log_odds()

# 2. Загрузка файла
file_path = "gen.fna"
record = next(SeqIO.parse(file_path, "fasta"))

sequence = record.seq[:1000000]

# 4 & 5. Поиск сайтов связывания со скором > 5.0
threshold = 5.0
print(f"Поиск хитов на прямой цепи (порог > {threshold}):")

# Метод search возвращает генератор пар (позиция, скор)
for position, score in pwm.search(sequence, threshold=threshold):
 print(f"Позиция: {position}, Скор: {score:.2f}, Последовательность: {sequence[position:position+m.length]}")

# 6. Поиск на обратной комплементарной цепи
print(f"\nПоиск хитов на обратной комплементарной цепи:")
reverse_sequence = sequence.reverse_complement()

for position, score in pwm.search(reverse_sequence, threshold=threshold):
 print(f"Позиция на RC: {position}, Скор: {score:.2f}")
```
